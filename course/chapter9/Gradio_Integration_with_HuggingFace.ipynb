```python
import gradio as gr

title = "GPT-J-6B"
description = "Gradio Demo for GPT-J 6B, a transformer model trained using Ben Wang's Mesh Transformer JAX. 'GPT-J' refers to the class of model, while '6B' represents the number of trainable parameters. To use it, simply add your text, or click one of the examples to load them. Read more at the links below."
article = "<p style='text-align: center'><a href='https://github.com/kingoflolz/mesh-transformer-jax' target='_blank'>GPT-J-6B: A 6 Billion Parameter Autoregressive Language Model</a></p>"

gr.load(
    "models/EleutherAI/gpt-j-6B",
    inputs=gr.Textbox(lines=5, label="Input Text"),
    title=title,
    description=description,
    article=article,
).launch()
```

This loading method is ideal for huge models like GPT-J or T0pp which require lots of RAM/computational power.

## Loading from Hugging Face Spaces ##

To load any Space from the Hugging Face Hub and recreate it locally, we can pass spaces/ to the Interface, followed by the name of the Space.

Let's now load a model that removes the background of an image!

```python
!pip install gradio rembg
```

```python
import gradio as gr
from rembg import remove

gr.Interface(remove, gr.Image(type="pil"), gr.Image(type="pil")).launch()
```

One of the cool things about loading demos from the Hub or spaces is that we can customize them by overriding any of the parameters. Here, we add a title and get it to work with a webcam instead:

```python
gr.Interface(remove, gr.Image(sources=["webcam"], type="pil"), gr.Image(type="pil")).launch()
```
